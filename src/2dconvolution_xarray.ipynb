```python
import xarray as xr
import numpy as np
from scipy.ndimage import convolve
from scipy.ndimage import gaussian_filter as gf
```

```python
# make up some random data in 2d
da=xr.DataArray(np.random.rand(20,20), dims=("x", "y"))
```

```python
# plot
da.plot(vmin=0, vmax=1)
```

```python
N=4
kernel2d = np.ones((N, N))/N**2
conv2d = lambda x: convolve(x, kernel2d, mode="wrap")
result2d = xr.apply_ufunc(conv2d, da[:, :])
```

```python
result2d.plot()
```

```python
N=3
kernel2d = np.ones((N, N))/N**2
result2d = xr.apply_ufunc(conv2d, da[:, :])
```

```python
result2d.plot()
```

```python
N=3
kernel2d = np.ones((N, N))/N**2
kernel2d = gf(kernel2d, sigma = 1, mode = 'constant' )
result2d = xr.apply_ufunc(conv2d, da[:,:])
result2d.plot()
```

```python
N=3
kernel2d = np.ones((N, N))/N**2
kernel2d = gf(kernel2d, sigma = 1, mode = 'reflect' )
result2d = xr.apply_ufunc(conv2d, da[:,:])
result2d.plot()
```

```python
N=3
kernel2d = np.ones((N, N))/N**2
kernel2d = gf(kernel2d, sigma = 1, mode = 'mirror' )
result2d = xr.apply_ufunc(conv2d, da[:,:])
result2d.plot()
```

```python
N=3
kernel2d = np.ones((N, N))/N**2
kernel2d = gf(kernel2d, sigma = 1, mode = 'wrap' )
result2d = xr.apply_ufunc(conv2d, da[:,:])
result2d.plot()
```

```python
N=3
kernel2d = np.ones((N, N))/N**2
kernel2d = gf(kernel2d, sigma = 1, mode = 'nearest' )
result2d = xr.apply_ufunc(conv2d, da[:,:])
result2d.plot()
```
